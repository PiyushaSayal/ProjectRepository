```python
import pandas as pd
```

```python
df = pd.read_csv('FinancialAnalysis.csv')
```

```python
df.head()
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])[' Net Income'].pct_change() * 100
```

```python
df.rename(columns={' Net Income': 'Net Income'}, inplace=True)
df.rename(columns={' Total Assets': 'Total Assets'}, inplace=True)
df.rename(columns={' Total Liabilities': 'Total Liabilities'}, inplace=True)
```

```python
df['Assets Growth (%)'] = df.groupby(['Company'])['Total Assets'].pct_change() * 100
df['Libailities Growth (%)'] = df.groupby(['Company'])['Total Liabilities'].pct_change() * 100
df['CFO Growth (%)'] = df.groupby(['Company'])['Cash Flow from Operating Activities.'].pct_change() * 100
```

```python
df.head()
```

```python
df = df.loc[:, ~df.columns.str.startswith('Unnamed')]
```

```python
df.head()
```

```python
df.fillna(0, inplace=True)
```

```python
df.head()
```
